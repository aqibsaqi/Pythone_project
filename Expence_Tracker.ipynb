```python
import csv
import json
import os
from datetime import datetime
```

```python
class ExpenseTracker:
    def __init__(self, csv_filename="expenses.csv", json_filename="expense_backup.json"):
        # Files ke naam hum ne class me save kar liye
        self.csv_file = csv_filename
        self.json_file = json_filename

        # Agar CSV file pehle se nahi bani hui, to header ke sath bana dete hain
        if not os.path.exists(self.csv_file):
            with open(self.csv_file, mode='w', newline='', encoding='utf-8') as file:
                writer = csv.writer(file)
                writer.writerow(["Date", "Title", "Amount"])
        # 1. Add Expense
    def add_expense(self):
        print("\n--- Add New Expense ---")
        title = input("Expense ka naam (e.g., Biryani, Petrol): ").strip()
        try:
            amount = float(input("Amount (Rs): "))
        except ValueError:
            print("❌ Galt amount! Sirf numbers likhein.")
            return

        date_str = datetime.now().strftime("%Y-%m-%d")

        # CSV file me data append (add) karna
        with open(self.csv_file, mode='a', newline='', encoding='utf-8') as file:
            writer = csv.writer(file)
            writer.writerow([date_str, title, amount])

        print(f" '{title}' (Rs {amount}) kamyabi se save ho gaya!")

    # 2. View Expenses
    def view_expenses(self):
        print("\n--- All Expenses ---")
        if not os.path.exists(self.csv_file):
            print(" Abhi tak koi expense record nahi mila.")
            return

        with open(self.csv_file, mode='r', encoding='utf-8') as file:
            reader = csv.reader(file)
            next(reader) # Header row ko skip karne ke liye

            count = 0
            for row in reader:
                if row:
                    print(f" {row[0]} |  {row[1]} |  Rs {row[2]}")
                    count += 1

            if count == 0:
                print(" File khali hai!")

    # 3. Save to CSV (File Handling me Add ke sath hi update hota hai, par isko hum manually refresh/verify ke liye use kar sakte hain)
    def save_to_csv(self):
        print("\n--- Save/Verify CSV File ---")
        if os.path.exists(self.csv_file):
            print(f" Aap ka saara data '{self.csv_file}' me mehfooz hai.")
        else:
            print("CSV file mojood nahi hai!")

    # 4. Backup to JSON
    def backup_to_json(self):
        print("\n--- Backup to JSON ---")
        if not os.path.exists(self.csv_file):
            print(" Backup ke liye koi data hi nahi hai!")
            return

        expenses_list = []
        # CSV se data read karna
        with open(self.csv_file, mode='r', encoding='utf-8') as file:
            reader = csv.DictReader(file)
            for row in reader:
                expenses_list.append({
                    "Date": row["Date"],
                    "Title": row["Title"],
                    "Amount": float(row["Amount"])
                })

        # JSON file me write karna
        with open(self.json_file, mode='w', encoding='utf-8') as file:
            json.dump(expenses_list, file, indent=4)

        print(f" Kamyabi se '{self.json_file}' me backup ban gaya hai!")

    # 5. Monthly Summary
    def monthly_summary(self):
        print("\n--- Monthly Summary ---")
        current_month = datetime.now().strftime("%Y-%m")
        total_spent = 0.0

        if not os.path.exists(self.csv_file):
            print(" Koi data nahi mila.")
            return

        with open(self.csv_file, mode='r', encoding='utf-8') as file:
            reader = csv.reader(file)
            next(reader)
            for row in reader:
                if row and row[0].startswith(current_month):
                    total_spent += float(row[2])

        print(f" Is mahine ({current_month}) ka kul kharcha: Rs {total_spent}")

    # 6. Search Expense
    def search_expense(self):
        print("\n--- Search Expense ---")
        query = input("Jo kharcha dhoondna hai uska naam likhein: ").lower().strip()
        found = False

        with open(self.csv_file, mode='r', encoding='utf-8') as file:
            reader = csv.reader(file)
            next(reader)
            for row in reader:
                if row and query in row[1].lower():
                    print(f" Mila: {row[0]} | {row[1]} | Rs {row[2]}")
                    found = True

        if not found:
            print(" Is naam ka koi expense nahi mila.")

    # 7. Delete Expense
    def delete_expense(self):
        print("\n--- Delete Expense ---")
        title_to_delete = input("Jis expense ko delete karna hai uska poora naam likhein: ").lower().strip()

        if not os.path.exists(self.csv_file):
            return

        rows_to_keep = []
        deleted = False

        # Header pehle add kar letay hain
        rows_to_keep.append(["Date", "Title", "Amount"])

        with open(self.csv_file, mode='r', encoding='utf-8') as file:
            reader = csv.reader(file)
            next(reader)
            for row in reader:
                if row:
                    if row[1].lower() == title_to_delete and not deleted:
                        # Pehla match jo milega usay skip (delete) kar dein ge
                        deleted = True
                        print(f" '{row[1]}' (Rs {row[2]}) ko delete kar diya gaya.")
                    else:
                        rows_to_keep.append(row)

        if not deleted:
            print(" Is naam ka koi expense nahi mila jo delete kiya ja sakay.")
            return

        # Ab bache hue data ko wapas file me write kar dete hain
        with open(self.csv_file, mode='w', newline='', encoding='utf-8') as file:
            writer = csv.writer(file)
            writer.writerows(rows_to_keep)
    def main_menu(self):
        while True:
            print("\n==============================")
            print("    EXPENSE TRACKER SYSTEM    ")
            print("==============================")
            print("1. Add Expense")
            print("2. View Expenses")
            print("3. Save / Verify CSV File")
            print("4. Backup to JSON File")
            print("5. View Monthly Summary")
            print("6. Search Expense")
            print("7. Delete Expense")
            print("8. Exit Program")
            print("==============================")

            choice = input("Apni choice select karein (1-8): ").strip()

            if choice == '1':
                self.add_expense()
            elif choice == '2':
                self.view_expenses()
            elif choice == '3':
                self.save_to_csv()
            elif choice == '4':
                self.backup_to_json()
            elif choice == '5':
                self.monthly_summary()
            elif choice == '6':
                self.search_expense()
            elif choice == '7':
                self.delete_expense()
            elif choice == '8':
                print("\n Shukriya! Program bhand ho raha hai.")
                break
            else:
                print(" Galt option! Dobara koshish karein.")
```

```python
if __name__ == "__main__":
    # Class ka object banaya aur project shuru kiya
    tracker = ExpenseTracker()
    tracker.main_menu()
```
