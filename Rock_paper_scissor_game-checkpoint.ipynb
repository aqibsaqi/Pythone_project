```python
'''
rock vs paper-->paper
rock vs scissors-->rock
paper vs scissors-->scissor
'''
import random
l=["rock","paper","scissors"]    
```

```python
u_score=0
c_score=0
while True:
    uc=int(input('''
    ===Game start===
    1 yes/start game
    2 No/exit
    '''))          
    if uc==1:
        for x in range(1,6):
            user_input=int(input('''
    ===Select any option====
    1- rock
    2- paper
    3- scissors
            '''))
            if user_input==1:
                uchoice="rock"
            elif user_input==2:
                uchoice="paper"
            elif user_input==3:
                uchoice="scissors"
            cchoice=random.choice(l)
            if uchoice==cchoice:
                print("user choice is :",uchoice)
                print("computer choice is:",cchoice)
                print("Game is Draw")
                u_score +=1
                c_score +=1
            elif (uchoice=="paper" and cchoice=="rock") or (uchoice=="rock" and cchoice=="scissors") or (uchoice=="scissors" and cchoice=="paper"):
                print("congratulation you  are win")
                u_score +=1
                print("user choice is :",uchoice)
                print("computer choice is:",cchoice)
            else:
                print(" computer are win" )
                c_score +=1
                print("user choice is :",uchoice)
                print("computer choice is:",cchoice)
            print("user score is :",u_score)
            print("computer score is :",c_score)
            
    else:
        break
                
                 
                 
                 
                 
```
